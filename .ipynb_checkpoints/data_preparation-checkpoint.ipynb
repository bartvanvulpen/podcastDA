## Data Preparation

```python
import json
import pandas as pd
```

```python
# load a podcast transcript from data path and print utterances/dialogue

path = 'podcast_data_no_audio/transcripts/podcasts-transcripts/0/5/show_05As0fQbe0p9CgcaJbek8n/2CJtQvWuxtxxgpjqcL72uP.json'
with open(path) as f:
    data = json.load(f)

for utterance_number, _ in enumerate(data['results']):  
    utterance_dict = data['results'][utterance_number]['alternatives'][0]
    try:
        utterance = utterance_dict['transcript']
    except KeyError:
        utterance = '';

    print(utterance)
```
